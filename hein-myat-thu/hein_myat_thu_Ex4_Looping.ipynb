Ex 1: Write a program that uses a while loop (not a for loop) to read through a string and print
the characters of the string one-by-one on separate lines.

```python
is_feeling = True

while is_feeling:

  print("--------------------------")
  emo = input("Enter whatever in your mind: ").replace(" ", "") # remove spaces from the input
  print("--------------------------")
  len_emo = len(emo) # get the length of the input (without spaces)
  count = 0

  if len_emo == 0: # ensure the user doesn't submit an empty input
    print("Please enter something!")
  else:
    while count < len_emo: # loop through each character in the input
      print(emo[count])
      count += 1 # move to the next character
    print("--------------------------")
    is_feeling = False # exit the main loop after successful input

"""
here, when the input is empty, if block runs. It will be looping until the user enters something.
then, when the input is valid, else block runs.
Inside that else block, there is another while loop which will loop through each character in the input.
Once the characters are printed, is_feeling = False is executed, which changes the loop condition to False,
and ends the main loop.
"""
```

Ex 2: A good program will make sure that the data its users enter is valid. Write a program that
asks the user for a weight and converts it from kilograms to pounds. Whenever the user
enters a weight below 0, the program should tell them that their entry is invalid and then ask
them again to enter a weight. [Hint: Use a while loop, not an if statement].

```python
is_kilos = True


while is_kilos:
  print("-------------------")
  weight = input("Enter weight in kilograms (Kg): ").strip() # remove leading/trailing whitespace
  # .strip() ensures that inputs like "   " (just spaces) become ""
  # plus, when user enter nothing or spaces and enter, it'll trigger ValueError
  # Since .strip() removes spaces, it contributes to checking for empty input using: if weight == "": in except block
  print("-------------------")

  try:
    # Try to convert the input to a float (to allow decimals)
    # If conversion fails (e.g. letters or empty string), it will raise a ValueError

    # Check if weight is positive
    kilo = float(weight)
    if kilo > 0:
      pound = kilo * 2.20462
      print(f"✏️{kilo} Kg = {pound:.2f} Pounds🧾")
      print("-------------------")
      is_kilos = False # exit the main loop after valid input
    else:
      # Handle zero or negative input
      print("📌WEIGHT can't be negative or zero❗️")
      print()
      print("📌Enter a valid weight❗️")

  except ValueError:
    # Handle non-numeric input
    print("📌Please enter number❗️")

    # If input was completely empty, ask user if they want to continue or exit the program
    # as i mention above, .strip() helps to convert empty string if the user hits just space
    if weight == "":
      print("-------------------")
      while True:
        # Ask user to confirm if they want to continue
        exit = input("Do you want to exit❓ (Y/N): ").strip().upper()
        print("-------------------")

        if exit == "Y":
          print("🔎Exiting.....❌")
          print("-------------------")
          print("🎉Have a nice day!🎉")
          print("-------------------")
          is_kilos = False # exit main loop
          break
        elif exit == "N":
          print("🔎Continue.....✅")
          break # break inner loop, and return to weight input (outer loop continues)
        else:
          # Handle invalid confirmation input
          print("📌Invalid input. Please type 'Y' or 'N'❗️")
          print("-------------------")

```

Ex 4: Write a program that asks the user for a string and a letter. Using a for loop, the
program should print the index of the first occurrence of that letter and a message if the
string does not contain the letter.

```python
# in this ex 4, I defined some methods so that my code can be more readable and robust
# I had to witness how logic is also important in ex 4 and ex 5
# what I've learnt in Discrete Math is really helpful here....
# I got some assistance from GPT, that doesn't mean I copied the code.....
# I just asked GPT what I want to know and the stuffs I don't understand
# I used it as a Guide or lecturer
# E.g., I know how to check a string whether it is aplha using .isalpha()
# but in some cases, I want to check each char in that string...so I asked GPT whether we can
# check each char in the string.....
# I used it in that way

def find_letter_index(text, letter): # this method is for finding letter
  for i in range(len(text)):
    if text[i] == letter: # when the letter we want find is found in the text
      return f"🎉The letter '{letter}' first occurs at index {i}.🎉" # it'll return this line


  return f"❌The letter '{letter}' is not in the string.❌" # if not, it'll return this line



def valid_text(text): # this method is for validation of the text
# I only want to allow alpha and space in the text like: Hello World
# num, special char, empty string, a mix of them wouldn't be allowed
  if text == "": # when user enters nothing or usser enters spaces, it'll return false
    return False
  for char in text: # checking each char
    if not (char.isalpha() or char.isspace()): # if each char is not either alpha or space, it'll return false
      return False # it returns flase, meaning that text input is not allowed
  return True # if the above conditions didn't meet, it'd return True, meaning allows the input (it's a valid input)

  """
  This is example explanation from GPT, i just wanna save it for revision

   Examples:

  "A" (a letter):
	•	char.isalpha() → True
	•	char.isspace() → False
	•	char.isalpha() or char.isspace() → True
	•	not True → False → OK (we don’t return False)

  From my prespective, what i've learned is that if condition will only triger the under line only if that condition is true
  that's why we use if not() here
  like in the example, when each char is met both condi or one condi, it'll give us Ture cuz we're using 'or' logical operator
  so if the condition is met, we have to allow that input and we don't need to return False
  here, if not() flip that True into False so that it will not return False
  So what we have to understand here is:
  if not (char.isalpha() or char.isspace()):
    return False

  return False will be triggered only when 'if True:'

  when conditions are met, it becomes like this:

  if not True: as conditions are met, we don't need to return False. Not to return, it has to be 'If False:'.

  to become 'if False:', not flip that True into False, then it skips return False and will trigger return True meaning input is valid.

  vice versa:

   when conditions are not met, it becomes like this:

   if not False: as conditions are not met, we need to return False. To return False, it has to be 'If True:'.

   to become 'if True:', not flip that False into True, then it triggers return False meaning input is invalid.

  """


while True:

  print("-------------------")
  text = input("Enter something in your mind: ").strip().lower()
  print("-------------------")
  if not valid_text(text):
    # here the reason of using 'not' is that
    # when valid_text() returns False (meaning invalid input), we have to print the below lines.
    # to trigger print(), need to be 'if True'
    # that's why 'False' (that is returned from valid_text()) is flipped into True using if not().
    # vice versa, when valid_text() return True, it will be flipped into 'False'
    # and becomes If Fasle:, therefore, it will skip the printing lines
    print("📌Invalid Input❗️")
    print("-------------------")
    print("📌Only Alphabets are allowed and Input must not be empty❗️")
    continue

  while True:

    letter = input("Enter a letter you want to find: ").strip().lower()
    print("-------------------")
    if len(letter) != 1 or not letter.isalpha(): # checking to ensure that the letter must be only one letter and must be Alpha
      print("📌Invalid Input❗️")
      print("-------------------")
      print("📌Please enter exactly ONE letter❗️")
      print("-------------------")
    else:
      break

  result = find_letter_index(text, letter)
  print(result)
  print("-------------------")
  break
```
